```python
%matplotlib inline
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
# La figure est l'objet parent de tout
fig = plt.figure()

# Le subplot est nécessaire pour afficher sur la figure
# la figure contient 1 ligne, 1 colonne et son prend l'ax n°1
ax = fig.add_subplot(1, 1, 1) 

"""
Autre façon de créer la figure t les subplots:
fig, axes = plt.subplots(2, 3) # 2 lignes, 3 colonnes
axes est un tableau qui contient chaque axe
axes[0, 1] # objet Axes 1ère ligne et 2ème colonne
"""

# linestyle pour le type de ligne
# marker pour le type de point
ax.plot([1,1,2], [1,2,3], linestyle='--', marker='o', color="g", label="Predictions")
```

```python
fig.clear()
fig = plt.figure()
y = range(1, 16)
x = [datetime(2017, 2, i) for i in y]

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y)

# rotation des labels sur les axes
# ne s'applique que sur les major
#ax.set_xticklabels(ax.get_xticks(), rotation=-90)

# major_locator sont les ticks dont le trai est supérieur
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=3))
#plt.setp(ax.get_xmajorticklabels(), rotation=90, fontsize=15)

# minor_locator sont les ticks dont le trai est inférieur
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%d-%m'))
#plt.setp(ax.get_xminorticklabels(), rotation=90, fontsize=10)

plt.xticks(rotation=70)
```

```python
fig.clear()
fig = plt.figure()
y = range(1, 16)
x = [datetime(2017, 2, 1, i) for i in y]

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%H'))
ax.xaxis.set_major_locator(mdates.HourLocator(byhour=range(12), interval=12))
ax.xaxis.set_minor_locator(mdates.HourLocator(byhour=range(2), interval=1))

plt.xticks(rotation=70)
```
